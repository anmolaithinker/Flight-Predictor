# Making of a new Dataset suitable for training
### The total number of flights combining all these spots in flight train dataset is more than or equal to 15 then it is good day else it is a bad day for paragliding. 

```python
# Function for calculating total number of flights
def cal(x):
    s = 0
    for i in range(1,289):
        s = s + x['Spot'+str(i)+' totalFlights']
    return s
flight_train['Total'] = flight_train.apply(cal , axis = 1)
```

```python
# Function for calculating whether the day is good day or bad day
def cal1(x):
    if(x['Total'] >= 15):
        return 1
    else:
        return 0
flight_train['Good_Bad'] = flight_train.apply(cal1 , axis = 1)
```

```python
# Making a new dataframe 
new_df = flight_train[['Day_Id' , 'Good_Bad']]
```

## Making a New Dataframe
- columns
<br>
    - S1DewPoint -> Average Station 1 DewPoint
    - S2DewPoint -> Average Station 2 DewPoint
    - S3DewPoint -> Average Station 3 DewPoint
    - S4DewPoint -> Average Station 4 DewPoint
    - S5DewPoint -> Average Station 5 DewPoint
<br>
    - S1 Press -> Station1 Pressure Difference
    - S2 Press -> Station2 Pressure Difference
    - S3 Press -> Station3 Pressure Difference
    - S4 Press -> Station4 Pressure Difference
    - S5 Press -> Station5 Pressure Difference
<br>
    - S1Temp -> Station1 Average Temp
    - S2Temp -> Station2 Average Temp
    - S3Temp -> Station3 Average Temp
    - S4Temp -> Station4 Average Temp
    - S5Temp -> Station5 Average Temp
<br>
    - S1 WS -> Station1 Average Wind Speed
    - S2 WS -> Station2 Average Wind Speed
    - S3 WS -> Station3 Average Wind Speed
    - S4 WS -> Station4 Average Wind Speed
    - S5 WS -> Station5 Average Wind Speed
<br>
    - S1 RH -> Station1 Approx. Relative Humidity
    - S2 RH -> Station2 Approx. Relative Humidity
    - S3 RH -> Station3 Approx. Relative Humidity
    - S4 RH -> Station4 Approx. Relative Humidity
    - S5 RH -> Station5 Approx. Relative Humidity    

```python
df = pd.DataFrame()
df_test = pd.DataFrame()
def dew(x,num):
    s = 0
    for i in range(1,46):
        s = s + x['Station'+str(num)+' Dew Point Height'+str(i)]
    return s/45

def Pressure(x,num):
    s = x['Station'+str(num)+' Pressure Height1'] - x['Station'+str(num)+' Pressure Height45']
    return s


def Temp(x,num):
    s = 0
    for i in range(1,46):
        s = s + x['Station'+str(num)+' Temperature Height'+str(i)]
    return s/45

def WS(x,num):
    s = 0
    for i in range(1,46):
        s = s + x['Station'+str(num)+' Wind Speed Height'+str(i)]
    return s/45

def RH(x,num):
    s = x['S'+str(num)+'Temp'] - x['S'+str(num)+'DewPoint']
    return s
```

# Model -> XGBoostClassifier (Score = 156)

```python
X = df.iloc[:, 0:25].values
y = df.iloc[:, 26].values
test = df_test.iloc[:,0:25].values
```

```python
X
```

```python
y
```

```python
# Fitting XGBoost to the Training set
from xgboost import XGBClassifier
classifier = XGBClassifier(n_estimators: 100, max_depth: 3)
classifier.fit(X , y)
```

### Applying GridSearch for fine tuning the parameters

```python
from sklearn.model_selection import GridSearchCV
parameters = [{'n_estimators': [100], 'max_depth': [3,4,5,6,7]}]
grid_search = GridSearchCV(estimator = classifier,
                           param_grid = parameters,
                           scoring = 'accuracy',
                           cv = 10,
                           n_jobs = -1,verbose=1)
grid_search = grid_search.fit(X, y)
best_accuracy = grid_search.best_score_
best_parameters = grid_search.best_params_
```
